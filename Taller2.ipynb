```python
#2 tonalidades#

imagen = Image.open(r'/content/drive/MyDrive/Mi carpetica/flores-bonitas.jpg')
width = imagen.size[0]
heigth = imagen.size[1]
for i in range(0,width):
  for j in range(0,heigth):
    data = imagen.getpixel((i,j))
    red=data[0]
    green=data[1]
    blue=data[2]
    pixel=int((red+green+blue)/3)
    if pixel >= 128:
        pixel = 63
    else:
      pixel=191
    imagen.putpixel((i,j), (pixel, pixel, pixel))
plt.imshow(imagen)
```
