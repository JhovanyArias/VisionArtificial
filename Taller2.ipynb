Convertir imagen a escala de grises Mostrar histograma para tonalidades de colores de 64 32 16 8 4 2

```python
#8 tonalidades#

imagen = Image.open(r'/content/drive/MyDrive/Mi carpetica/flores-bonitas.jpg')
width = imagen.size[0]
heigth = imagen.size[1]
for i in range(0,width):
  for j in range(0,heigth):
    data = imagen.getpixel((i,j))
    red=data[0]
    green=data[1]
    blue=data[2]
    pixel=int((red+green+blue)/3)
    if pixel <= 31:
      pixel=15
    elif pixel >=32 and pixel <= 63:
        pixel = 47
    elif pixel >= 64 and pixel <= 95:
        pixel = 79
    elif pixel <= 96 and pixel <= 127:
      pixel = 111
    elif pixel >=128 and pixel <= 159:
        pixel = 143
    elif pixel >= 160 and pixel <= 191:
        pixel = 175
    elif pixel >= 192 and pixel <= 223:
        pixel = 207
    else:
      pixel=239
    imagen.putpixel((i,j), (pixel, pixel, pixel))
plt.imshow(imagen)
```
